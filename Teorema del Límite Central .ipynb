```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
grades = [10,9,8,7,10,7,6,10,8,9,8,10,5,7,8,9,10,5,5,7,6,8,9,10,5,10,9,8,9,7,8,8,8,7,8,4,5,8,6,4,5,6,8]
grades = np.array(grades)
len(grades)
```

```python
grades.mean()
```

```python
x = np.random.choice(grades,10)
x
```

```python
means = []


for i in range(5000):
    x = 0
    mu = 0
    
    x = np.random.choice(grades,10)
    mu = x.mean()
    means.append(mu)
    
means = np.array(means)
```

```python
plt.figure(figsize=(8,6))
plt.hist(means, bins=20, ec="black", color="lightblue")
plt.title("Distribución de medias de calificaciones de Matemáticas")
plt.xlabel("Calificaciones")
plt.ylabel("Frecuencia")
plt.grid(alpha=0.4)
plt.show()

print("Calificación Media",means.mean())
```

```python
#Curva normal
import scipy.stats as ss

mu = means.mean()
sd = means.std()

x_min = mu - 4*sd
x_max = mu + 4*sd

x = np.linspace(x_min, x_max, 100)
y = ss.norm.pdf(x, mu, sd)

plt.figure(figsize=(8,6))
plt.plot(x, y, color="blue", alpha=0.7)
plt.title("Distribución de medias de calificaciones de Matemáticas")
plt.xlabel("Calificaciones")
plt.ylabel("Frecuencia")
plt.grid(alpha=0.4)
plt.show()
```
