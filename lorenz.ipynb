# Efecto Mariposa

En 1952 en la novela de ‚ÄúThe Thunder‚Äù de Ray Bradbury (mismo autor de Fahrenheit 451)
aparece lo que se conoce como ‚Äúel efecto mariposa‚Äù que ilustra c√≥mo es que en
determinadas circunstancias una peque√±a variaci√≥n en las condiciones iniciales puede tener
grandes consecuencias en el mediano o largo plazo.

En 1963 Edward Lorenz, padre de la teor√≠a del caos, desarrollo una soluci√≥n para un modelo
de oscilaci√≥n atmosf√©rica conocida como el atractor de Lorenz. Casualmente, se descubri√≥
que para una determinada combinaci√≥n de ùúé (n√∫mero de Prantdl) y de ùúå (n√∫mero de
Rayleigh), las oscilaciones del atractor tienen un comportamiento ca√≥tico y las trayectorias
se asemejan bastante a las alas de una mariposa.


### Ecuaciones  de Lorenz
$x' = \sigma (y-x),$ <br>
$y' = x(\rho-z)-y$, <br>
$z' = xy-\beta z$

```python
#MANU1_U3_AC
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import integrate as spi 
import numpy as np
```

```python
def rhs(u,t,beta,rho,sigma):
    x,y,z=u
    return [sigma*(y-x), x(rho-z)-y, x*y-beta*z]
```

```python
sigma = 10
beta = 8/3
rho1 = 29
rho2 = 28

u01 = [1,1,1]
u02 = [1,1,1]

t = np.linspace(0,50,10001)
```

```python
u1 = spi.odeint(rhs,u01,t,args=(beta,rho1,sigma))
u2 = spi.odeint(rhs,u02,t,args=(beta,rho2,sigma))
```

```python
x1,y1,z1=u1[:,0],u1[:,1],u1[:,2]
x2,y2,z2=u2[:,0],u2[:,1],u2[:,2]
```

```python
plt.ion()
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x1,y1,z1,'blue', alpha=0.5)
ax.plot(x2,y2,z2,'r:', alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Lorenz Equation')
```
