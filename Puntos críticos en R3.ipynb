```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(-4,4)
x,y = np.meshgrid(t,t)
f = lambda x,y: y**3 + 3*(x**2)*y - 6*(x**2) - 6*(y**2) + 2

fig = plt.figure(figsize=(22,8))
ax = fig.add_subplot(121, projection='3d')
ax.set_title("$f(x,y,z)$", fontsize=20)

ax.plot_surface(x,y,f(x,y), color="lightblue", alpha=0.8)
ax.scatter(-2,2,f(-2,2),color="black")
ax.scatter(2,2,f(2,2),color="black")
ax.scatter(0,0,f(0,0),color="black")



ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)

ax.grid()
plt.show()
```
