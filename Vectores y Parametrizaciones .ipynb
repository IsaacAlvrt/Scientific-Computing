```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
#vector A,B,C
A = [-1,-2]
B = [-3, 4]
C = [-4, 2]

head_length = 0.2

#magnitud de A, B, C
mag_A = np.sqrt(A[0]**2 + A[1]**2)
mag_B = np.sqrt(B[0]**2 + B[1]**2)
mag_C = np.sqrt(C[0]**2 + C[1]**2)

#arrow
Adx = (A[0] / mag_A) * (mag_A - head_length)
Ady = (A[1] / mag_A) * (mag_A - head_length)

Bdx = (B[0] / mag_B) * (mag_B - head_length)
Bdy = (B[1] / mag_B) * (mag_B - head_length)

Cdx = (C[0] / mag_C) * (mag_C - head_length)
Cdy = (C[1] / mag_C) * (mag_C - head_length)



#Gráfica
plt.figure(figsize=(6,6))

#A
plt.plot(-1,-2,  label="$ \overrightarrow{v} =[-1,-2]$", color="orange")
plt.arrow(0,0, Adx, Ady, head_width=0.1, head_length=head_length, color="orange")

#B
plt.plot(-3, 4,  label="$ \overrightarrow{u} =[-3,4]$", color="green")
plt.arrow(0,0, Bdx, Bdy, head_width=0.1, head_length=head_length, color="green")

#C
plt.plot(-4, 2,  label="$ \overrightarrow{w} =[-4,2]$", color="blue")
plt.arrow(0,0, Cdx, Cdy, head_width=0.1, head_length=head_length, color="blue")

plt.title("Vectores en $\mathbb{R}^2$", fontsize=17)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.axis([-5,1,-2.5,4.5])
plt.axhline(color="gray")
plt.axvline(color="gray")
plt.legend()
plt.grid(alpha=0.6)
```

###### La parametrización para una linea que pasa por los puntos $A,B$, esta dada por:
###### $\langle x,y,z \rangle = A + t(B-A)$

```python
%matplotlib inline


t = np.linspace(-2,2)

xt = -1 + 4*t
yt = 2 - 5*t
zt = 5 - 5*t


fig = plt.figure(figsize=(18,7))
ax = fig.add_subplot(121, projection='3d')
ax.set_title("Parametrización", fontsize=17)


#Puntos
ax.scatter(-1, 2,5, label="A", marker="o", color="blue")
ax.scatter(3, -3, 0, label="B", marker="D", color="red")

#Recta
ax.plot(xt,yt,zt, color="orange")

ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)

ax.legend()
ax.grid()
plt.show()
```

###### La parametrización para un segmento que pasa por los puntos $P,Q$, esta dada por:
###### $\langle x,y,z \rangle = P + t(\overrightarrow{PQ})$, con $t \in [a,b]$

```python
P=np.array([-1,3,2])
Q=np.array([-8,1,5])
PQ = Q-P
PQ
```

```python
t=np.linspace(0,1,50)
x = -1 -7*t
y = 3 -2*t
z = 2 + 3*t


fig = plt.figure(figsize=(18,7))
ax = fig.add_subplot(121, projection='3d')
ax.set_title("Parametrización", fontsize=17)


#Puntos
ax.scatter(P[0],P[1],P[2], label="P", marker="o", color="blue")
ax.scatter(Q[0],Q[1],Q[2], label="Q", marker="D", color="red")

#Recta
ax.plot(x,y,z, color="orange")

ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)

ax.legend()
ax.grid()
plt.show()
```

## $h(x) = (x, x^2, \frac{\sin{x}}{x}) $

```python
t = np.linspace(-1,1,100)
x1 = t
x2 = t**2
x3 = np.sin(t)/t


fig = plt.figure(figsize=(22,8))
ax = fig.add_subplot(121, projection='3d')
ax.set_title("$h(x)= (x, x^2, \sin{x}/{x})$", fontsize=20)

#h(x)
ax.plot(x1,x2,x3, color="blue")

#axis
ax.plot([-1,1],[0,0],[0,0], ":", alpha=0.8, color="black")
ax.plot([0,0],[-1,1],[0,0], ":", alpha=0.8, color="black")
ax.plot([0,0],[0,0],[0,2], ":", alpha=0.8, color="black")

ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)

ax.grid()
plt.show()
```
