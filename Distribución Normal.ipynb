<h1 style="color: #14549C">Distribución Normal</h1>

### Introducción
Es el modelo de distribución de probabilidades más utilizado. Se dice que una variable aleatoria continua $X$ está normalmente distribuida si su función de densidad de probabilidad tiene la siguiente forma:
> <h2>$f(x)= \frac{1}{ \sigma \sqrt{2}} e^{- \frac{(x-\mu)^2}{2\sigma^2}} $</h2>

### Propiedades

    <ul>

        <li>La distribución normal depende de dos parámetros, μ y σ2. Por tanto, una vez especificados, podemos encontrar la probabilidad de que X esté dentro de cierto intervalo mediante la función de densidad de la distribución normal.
        </li>
    </ul>
</div>

Las áreas bajo la función de densidad normal correspondientes a $P(a ≤ X ≤ b)$ requieren
la evaluación de la integral:
> <h2> $\int_{a}^{b} \frac{1}{ \sigma \sqrt{2}} e^{- \frac{(x-\mu)^2}{2\sigma^2}} dx$ </h2>

Desafortunadamente, esta integral no puede ser resuelta de forma analítica (o por lo menos no con los métodos de integración comunes) por lo que se ha recurrido al desarrollo de una función de densidad de probabilidad estandar con $\mu=0$ y $\sigma=1$ que se resuelve mediante métodos numéricos.

> Estandarización de una variable aleatoria continua $X$ con media $\mu$ y desviación estándar $\sigma$
>
> <h2>$Z=\frac{x-\mu}{\sigma} $</h2>
>
>
>Función de densidad de probabilidad normal estandarizada
>
> <h2> $ f(Z)=\frac{1}{\sqrt{2\pi}} e^{- \frac{Z^2}{2}} $ </h2>

Para expresar que una variable aleatoria continua se distribuye de forma normal, se usa la notación:
$X ∼ N(μ, σ^2)$

```python
#Librerias usadas
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
import seaborn as sns
```

## Ejemplo
En una carrera, la llegada en segundos a la meta media fue 80 y la desviación estándar fue 10. Calcule las siguientes probabilidades mediante estandarización:

```python
#Datos
mu = 80
std = 10
```

###### a) La probabilidad de un corredor llegue en 100 segundos o menos a la meta: 

```python
x = 100

#cdf = comulative distribution function, calcula la probablidad de 0 a x
Px = ss.norm.cdf(x, mu, std)

print("La probabilidad de llegar en 100 seg o menos es:", round(Px,5))
```

###### b) Grafica el resultado del inciso a

```python
#Datos para graficar
mu = 80
sigma = 10

x_min = mu - (4*sigma)
x_max = mu + (4*sigma)

x_values = np.linspace(x_min, x_max, 10000) 
y_values = ss.norm.pdf(x_values, mu, sigma) #pdf = probility density function

#Gráfica
plt.figure(figsize=(8,5))
plt.plot(x_values, y_values, color="orange")
plt.title("Distribución de Probabilidades, $\mu=80, \sigma=10$")
plt.fill_between(x_values, y_values, 
                 where=[(x_values<100) for x_values in x_values], 
                 alpha=0.7, color="lightblue",
                label="$P(X \leq 100)$")
plt.xlabel("Segundos")
plt.legend()
plt.grid()
plt.show()
```

###### c) Calcula la probabilidad de que un corredor llegue entre 70 y 90 segundos

```python
#Probabilidad de lleguar en 70 seg o menos
x1 = 70
Px1 = ss.norm.cdf(x1, mu, std)

#Probabilidad de llegar en 90 seg o menos
x2 = 90
Px2 = ss.norm.cdf(x2, mu, std)

#Probabilidad de llegar entre 70 y 90 segs
Px = round(Px2 - Px1,5)
print("La probabilidad de llegar entre 70 y 90 segundos es de:", Px)
```

###### d) Grafica el resultado del inciso anterior

```python
#Datos para graficar
mu = 80
std = 10

x_min = mu - 4*std
x_max = mu + 4*std

x = np.linspace(x_min, x_max, 10000)
y = ss.norm.pdf(x, mu, std)

#Grafica
plt.figure(figsize=(8,5))
plt.plot(x,y, color="orange")
plt.title("Distribución de Probabilidades, $\mu=80, \sigma=10$")
plt.fill_between(x, y, 
                 where=[(x>70) and (x<90) for x in x], 
                 alpha=0.7, color="lightblue",
                label="$P(70 \leq X \leq 90)$")
plt.xlabel("Segundos")
plt.legend()
plt.grid()
plt.show()
```

###### e) Calcula la probabilidad de que un corredor llegue entre 85 y 95 segundos

```python
#probabilidad de que llegue en 85 segundos o menos
Px1 = ss.norm.cdf(85, mu, std)

#probabilidad de que llegue en 95 segundos o menos
Px2 = ss.norm.cdf(95, mu, std)

#probabilidad de que llegue entre 85 y 95 segundos
Px = round(Px2 - Px1, 5)
print("La probabilidad de llegar entre 85 y 95 segundos es de:", Px)
```

###### d) Grafica el resultado del inciso anterior

```python
#Datos para graficar
x_min = mu - 4*std
x_max = mu + 4*std

x = np.linspace(x_min, x_max, 10000)
y = ss.norm.pdf(x, mu, std)

#Grafica
plt.figure(figsize=(8,5))
plt.plot(x,y, color="orange")
plt.title("Distribución de Probabilidades, $\mu=80, \sigma=10$")
plt.fill_between(x, y, 
                 where=[(x>85) and (x<95) for x in x], 
                 alpha=0.7, color="lightblue",
                label="$P(85 \leq X \leq 95)$")
plt.xlabel("Segundos")
plt.legend()
plt.grid()
plt.show()
```
