```python
#Librerias
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Datos: Azucar (gramos) por tasa de 77 cereales 
data = pd.read_excel('Datasets/cereal.xlsx')[['name', 'sugars']]
data.head()
```

```python
#Estadisticas basicas
data.describe()
```

```python
#Histograma
data.hist(color='lightblue', ec='gray', bins=15)
plt.title('Azucar gr')
plt.xlabel('Gramos')
plt.ylabel('Frecuencia')
plt.grid(alpha=0.3)
plt.show()
```

```python
#Datos con azucar igual o mayor a 10 gr por cada 30 gr
od = data['sugars'] >= 10 
od.value_counts()
```

```python
#Muestra n=30
sample = data.sample(30)
sample
```

```python
#Estadisticas de la muestra
sample.describe()
```

```python
# Intervalo de confianza para la media (mu) para el azucar
n = 77
xbar = data.mean()
sigma = data.std()
z = 1.645 # nivel de confianza 90%

mu_min = xbar - z* (sigma / np.sqrt(n))
mu_max = xbar + z* (sigma / np.sqrt(n))

print(mu_min, mu_max)
```
